```python
# importar librerías necesarias
from kmeans import MetodoKMeans
from dbscan import MetodoDBSCAN
from Agglomerative import MetodoAgglomerative
from Gaussian import MetodoGaussian

from sklearn.datasets import load_iris, load_digits, load_wine, load_breast_cancer, load_diabetes, load_linnerud
```

```python
# importar base de datos
iris_data = load_iris()
```

# K-Means

```python
# llamar MetodoKMeans
metodo_km = MetodoKMeans(iris_data)

# Imprimir tabla resumen
print(metodo_km.tabla_resumen())
```

```python
# Imprimir silhouette score para diferentes clústeres
print(metodo_km.mejor_cluster())

# Se obtiene que la cantidad recomendada de clústeres es 2
```

```python
# Obtener tabla con el clúster perteneciente
clustered_data = metodo_km.clusters(2)
print(clustered_data.head())
```

```python
#  Graficar resultados en un pairplot
metodo_km.graficar(2)
```

```python
#  Obtener silhouette score para el clúster escogido
silueta_kmeans = metodo_km.silhouette(2)
print(silueta_kmeans)
```

# DBSCAN

```python
# llamar MetodoDBSCAN
metodo_db = MetodoDBSCAN(iris_data)

# Imprimir tabla resumen
print(metodo_db.tabla_resumen())
```

```python
# Imprimir silhouette score para diferentes clústeres
print(metodo_db.mejor_cluster())

# Se obtiene que la cantidad recomendada de eps es 0.9 y de min_samples es 2
```

```python
# Obtener tabla con el clúster perteneciente
clustered_data_dbscan = metodo_db.clusters(eps=0.9, min_samples=2)  
print(clustered_data_dbscan.head())
```

```python
#  Graficar resultados en un pairplot
metodo_db.graficar(eps=0.9, min_samples=2)
```

```python
#  Obtener silhouette score para el clúster escogido
silueta_dbscan = metodo_db.silhouette(0.9, 2)
print(silueta_dbscan)
```
