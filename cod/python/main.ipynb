```python
from kmeans import MetodoKMeans
from dbscan import MetodoDBSCAN
from Agglomerative import MetodoAgglomerative
from Gaussian import MetodoGaussian

from sklearn.datasets import load_iris, load_digits, load_wine, load_breast_cancer, load_diabetes, load_linnerud
```

```python
iris_data = load_iris()
```

# K-Means

```python
metodo_km = MetodoKMeans(iris_data)

print(metodo_km.tabla_resumen())

clustered_data = metodo_km.clusters(3, 42)
print(clustered_data.head())

metodo_km.graficar(3, 42)

silueta_kmeans = metodo_km.silhouette(3, 42)
print(silueta_kmeans)
```

# DBSCAN

```python
metodo_db = MetodoDBSCAN(iris_data)
print(metodo_db.tabla_resumen())

clustered_data_dbscan = metodo_db.clusters(eps=0.5, min_samples=5)
print(clustered_data_dbscan.head())

metodo_db.graficar(eps=0.5, min_samples=5)
    
silueta_dbscan = metodo_db.silhouette(0.5, 5)
print(silueta_dbscan)
```

# Agglomerative

```python
metodo_ag = MetodoAgglomerative(iris_data)

print(metodo_ag.tabla_resumen())

clustered_data = metodo_ag.clusters()
print(clustered_data.head())

metodo_ag.graficar()

silueta_agglo = metodo_ag.silhouette()
print(silueta_agglo)
```

# Gaussian Mixture

```python
metodo_gm = MetodoGaussian(iris_data)
print(metodo_gm.tabla_resumen())

clustered_data_gm = metodo_gm.clusters(3, 42)
print(clustered_data_gm.head())

metodo_gm.graficar(3, 42)

silueta_gm = metodo_gm.silhouette(3, 42)
print(silueta_gm)
```
